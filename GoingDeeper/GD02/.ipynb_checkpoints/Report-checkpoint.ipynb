+각 모델들의 성능을 단어장의 크기에 따라서 얼마나 차이가 있는지 비교 실험.

비교할 단어장의 크기


- 머신러닝 모델에는 TF-IDF로 벡터화 시킨 데이터를, 딥러닝에는 뉴스데이터 벡터 그대로 입력으로.. 
- 성능을 확인하는 지표로 정확도가 아닌 가중치가 적용된 F1-score(weighted)를 사용

### 결론

####  Naive Bayes 모델과 트리 기반의 모델의 경우, vocab_size가 커질수록 성능이 떨어지는 현상  
- 분류 할 수 있는 가지의 수가 많아지기 때문에, 옳바르게 분류할 확률이 떨어짐..?

#### Logistic Regression 모델과 Support Vector 모델의 경우, vocab_size가 커질수록 성능이 개선되는 현상  
- 1번이냐 아니냐 접급으로 인해 계산량은 많지만, 모든 클래스에 대해서 접근하기 때문...?

#### 딥러닝 모델의 경우, vocab_size가 커질수록 성능이 상승함.  
- 임베딩시킬 단어의 수가 늘어나면서, 경우의 수가 증가?
